```python
def softmax(vec, beta = 1):
#PARAMETRE ENTREE : NOTES FORMAT LISTE/ARRAY
#PARAMETRE ENTREE : BETA
#PARAMETRE SORTIE : PROBABILITES FORMAT LISTE
  #Retourne [ P1, P2 ]
    exponential = np.exp([element * beta for element in vec])
    probabilities = exponential / np.sum(exponential)
    return probabilities
```

```python
def comptage(a,b,c):
#COMPARE SIMPLEMENT A B ET C
#PARAMETRE SORTIE : 1 représente le plus petit
#                   0 n'est pas le plus petit
  if a < b and a < c:
    return 1,0,0
  elif b < a and b < c:
    return 0,1,0
  elif c < a and c < b:
    return 0,0,1
```

```python
def BIC(LL,N,parameter):
#PARAMETRE ENTREE : LOG-LIKELIHOOD FORMAT LISTE
#PARAMETRE ENTREE : NOMBRE DE CHOIX (82)
#PARAMETRE ENTREE : NOMBRE DE PARAMETRES LIBRES DU MODELE
#PARAMETRE SORTIE : CRITERE BAYESIEN FORMAT LISTE
    BIC = []

    for i in range(len(LL)):
        BIC.append(-2*LL[i]+parameter*log(N))

    return BIC
```

```python
def lineaire(a,n):
#PARAMETRE ENTREE : COEFFICIENT DIRECTEUR
#PARAMETRE ENTREE : NOMBRE DE PERSONNE POUR LE DESSERT EN QUESTION
  return (a/2)*n-a
```

```python
def NON_LINEAIRE(a,n):
#PARAMETRE ENTREE : COEFFICIENT DIRECTEUR
#PARAMETRE ENTREE : NOMBRE DE PERSONNE POUR LE DESSERT EN QUESTION
  if n == 4:
    return a
  else:
    return 0
```

```python
def generate_data_uniform():
    
    #INITIALISATION
    sains = {'sain':[], 'valeur':[]}
    nonsains = {'nonsain':[],'valeur':[] }
    
    #EXEMPLE DE DESSERTS
    sains['sain'] = ['Smoothie', 'Crème dessert à la vanille', 'Salade de Fruits', 'Yaourt sucré', 'Orange', 'Ile flottante', 'Liégeois']
    nonsains['nonsain'] = ['Quatre-Quart', 'Sorbet', 'Glace', 'Charlotte aux Fruits', 'Sundae', 'Tarte poires amandes', 'Mille-Feuille']
    
    #GENERATION DES NOTES
    nonsains['valeur'] =  np.random.uniform(low=0, high=7, size=7)
    sains['valeur'] = np.random.uniform(low=0, high=7, size=7)
    
    #VERIFICATION QUE TOUT EST BIEN DES INT
    for i in range(len(sains['valeur'])):
        sains['valeur'][i] = int(sains['valeur'][i])
        nonsains['valeur'][i] = int(nonsains['valeur'][i])
        
    return sains, nonsains
```

```python
def compare(fight, b):
#PARAMETRE ENTREE : NOTES DES DESSERTS 1 ET 2
#PARAMETRE ENTREE : BETA
#PARAMETRE SORTIE : CHOIX EN FONCTION DES PROBABILITES POUR LE DESSERT 0 ( gauche ) OU LE DESSERT 1 ( droite )
    probabilities = softmax(fight, beta = b)
    liste = [0,1]
    return random.choices(liste, weights = probabilities, k = 1)[0]
```

```python
def compare_lineaire(fight,b,a,i1,i2):
#PARAMETRE ENTREE : NOTES DES DESSERTS 1 ET 2
#PARAMETRE ENTREE : BETA
#PARAMETRE ENTREE : ALPHA
#PARAMETRE ENTREE : INFLUENCE POUR LE DESSERT 1
#PARAMETRE ENTREE : INFLUENCE POUR LE DESSERT 2
#PARAMETRE SORTIE : CHOIX EN FONCTION DES PROBABILITES POUR LE DESSERT 0 ( gauche ) OU LE DESSERT 1 ( droite )
    fight = [fight[0]+lineaire(a,i1),fight[1]+lineaire(a,i2)]
    probabilities = softmax(fight, beta = b)
    liste = [0,1]
    return random.choices(liste, weights = probabilities, k = 1)[0]
```

```python
def compare_declenchement(fight,b,d,i1,i2):
#PARAMETRE ENTREE : NOTES DES DESSERTS 1 ET 2
#PARAMETRE ENTREE : BETA
#PARAMETRE ENTREE : DECLENCHEMENT
#PARAMETRE ENTREE : INFLUENCE POUR LE DESSERT 1
#PARAMETRE ENTREE : INFLUENCE POUR LE DESSERT 2
#PARAMETRE SORTIE : CHOIX EN FONCTION DES PROBABILITES POUR LE DESSERT 0 ( gauche ) OU LE DESSERT 1 ( droite )
    fight = [fight[0]+NON_LINEAIRE(d,i1),fight[1]+NON_LINEAIRE(d,i2)]
    probabilities = softmax(fight, beta = b)
    liste = [0,1]
    return random.choices(liste, weights = probabilities, k = 1)[0]
```

```python
def ESSAI(BETA_SIMULATION, ALPHA_SIMULATION, D_SIMULATION, modele = 'no'):
#PARAMETRE ENTREE : BETA SIMULE
#PARAMETRE ENTREE : ALPHA SIMULE
#PARAMETRE ENTREE : D SIMULE
#PARAMETRE ENTREE : MODELE QUE L'ON SOUHAITE

#PARAMETRE SORTIE : L'ENSEMBLE DES COMPARAISONS ENTRE DESSERTS (noms)
#PARAMETRE SORTIE : L'ENSEMBLE DES CHOIX DE NOTRE SIMULATION
#PARAMETRE SORTIE : L'ENSEMBLE DES COMPARAISONS ENTRE DESSERTS (valeurs)
#PARAMETRE SORTIE : L'ENSEMBLE DES COMPARAISONS ENTRE DESSERTS (probabilités)
#PARAMETRE SORTIE : L'ENSEMBLE DES INFLUENCES


  #INITIALISATION
  
    duel_valeur = []
    duel_proba = []
    CHOIX_SIMULATION = []
    LL = []

    sains, nonsains = generate_data_uniform()
    data = pd.DataFrame(create_list_experiment([sains['sain'],nonsains['nonsain']])).drop(['times'], axis = 1)
    duel = data['pair']
    autrui = data['choices']
    sains = pd.DataFrame(sains).set_index(['sain'])
    nonsains = pd.DataFrame(nonsains).set_index(['nonsain'])
    desserts = pd.concat([sains,nonsains],axis = 0)

  #Modele sans influence
    if modele == 'no':
        INFLUENCE = []
        for i in range (len(autrui)):
            INFLUENCE.append(Counter(autrui[i]))
            
        for i in range (len(duel)):
            fight = [desserts.loc[duel[i][0]][0],desserts.loc[duel[i][1]][0]]
            duel_valeur.append(fight)
            duel_proba.append(softmax(fight,beta = BETA_SIMULATION))
            CHOIX_SIMULATION = duel[i][random.choices([0,1], weights = duel_proba[i], k = 1)[0]]
            
        return duel, CHOIX_SIMULATION, duel_valeur, duel_proba, INFLUENCE


  #Modèle linéaire
    elif modele == 'linear':
        INFLUENCE = []
        for i in range (len(autrui)):
            INFLUENCE.append(Counter(autrui[i]))
            
        for i in range (len(duel)):
            fight = [desserts.loc[duel[i][0]][0]+lineaire(ALPHA_SIMULATION,INFLUENCE[i][duel[i][0]]),desserts.loc[duel[i][1]][0]+lineaire(ALPHA_SIMULATION,INFLUENCE[i][duel[i][1]])]
            duel_valeur.append(fight)
            duel_proba.append(softmax(fight,beta = BETA_SIMULATION))
            CHOIX_SIMULATION = duel[i][random.choices([0,1], weights = duel_proba[i], k = 1)[0]]
            
        return duel, CHOIX_SIMULATION, duel_valeur, duel_proba, INFLUENCE

  #Modèle déclenchement
    elif modele == 'triggering':
        
        INFLUENCE = []
        for i in range (len(autrui)):
            INFLUENCE.append(Counter(autrui[i]))
        
        for i in range (len(duel)):
            fight = [desserts.loc[duel[i][0]][0]+NON_LINEAIRE(D_SIMULATION,INFLUENCE[i][duel[i][0]]),desserts.loc[duel[i][1]][0]+NON_LINEAIRE(D_SIMULATION,INFLUENCE[i][duel[i][1]])]
            duel_valeur.append(fight)
            duel_proba.append(softmax(fight,beta = BETA_SIMULATION))
            CHOIX_SIMULATION = duel[i][random.choices([0,1], weights = duel_proba[i], k = 1)[0]]
            
        return duel, CHOIX_SIMULATION, duel_valeur, duel_proba, INFLUENCE
```

```python
def LIKELIHOOD(DESSERT, CHOIX, VALEURS,PROBA_ORIGINELLE, BETA = 0, ALPHA = 0, D = 0,INFLUENCE = 0, modele = 'no', option ='beta'):
#PARAMETRE ENTREE : LISTE DES DESSERTS
#PARAMETRE ENTREE : CHOIX SIMULATION
#PARAMETRE ENTREE : PROBABILITES

#PARAMETRE SORTIE : LE PARAMETRE SOUHAITE
#PARAMETRE SORTIE : LE LOG LIKELIHOOD ASSOCIE

    CHOIX = []
    LL = []
    LL_ALPHA = []
    LL_D = []

    if modele == 'no':
        for round in range(len(DESSERT)):
            place = compare(VALEURS[round], BETA)
            LL.append(math.log(PROBA_ORIGINELLE[round][place]))
        
        return BETA, sum(LL)

    elif modele == 'linear':
        if option == 'beta':
            for round in range(len(DESSERT)):
                if INFLUENCE[round][DESSERT[round][0]] == 2:
                    place = compare(VALEURS[round], BETA)
                    LL.append(math.log(PROBA_ORIGINELLE[round][place]))
                    
            return BETA, sum(LL)

        elif option == 'alpha':
            for round in range(len(DESSERT)):
                place = compare_lineaire(VALEURS[round], BETA, ALPHA,INFLUENCE[round][DESSERT[round][0]],INFLUENCE[round][DESSERT[round][1]])
                LL_ALPHA.append(math.log(PROBA_ORIGINELLE[round][place]))
            
            return ALPHA, sum(LL_ALPHA)

    elif modele == 'triggering':

        if option == 'beta':
            for round in range(len(DESSERT)):
                if INFLUENCE[round][DESSERT[round][0]] == 2:
                    place = compare(VALEURS[round], BETA)
                    LL.append(math.log(PROBA_ORIGINELLE[round][place]))
                    
            return BETA, sum(LL)
      
        elif option =='triggering':
            for round in range(len(DESSERT)):
                place = compare_declenchement(VALEURS[round], BETA, D,INFLUENCE[round][DESSERT[round][0]],INFLUENCE[round][DESSERT[round][1]])
                CHOIX.append(DESSERT[round][place])
                LL_D.append(math.log(PROBA_ORIGINELLE[round][place]))
            
            return D, sum(LL_D)
```

```python
def simulation(B_SIMULATION = 0, A_SIMULATION = 0, D_SIMULATION = 0, MODELE = 'no'):


    if MODELE == 'no':
    #GENERATION POUR BETA SUIVANT UNE LOI UNIFORME
        DESSERT, CHOIX, VALEURS, PROBA, INFLUENCE = ESSAI(B_SIMULATION, 0, 0, modele = MODELE)

    #CONSERVATION DES DONNEES
        BANQUE = {'BETA': [], 'LL': []}
    #RECHERCHE
        for TEST in RECHERCHE:
            BETA, LL = LIKELIHOOD(DESSERT,CHOIX, VALEURS, PROBA, TEST )
            BANQUE['BETA'].append(BETA)
            BANQUE['LL'].append(LL)

        return BANQUE, B_SIMULATION

    elif MODELE == 'linear':

    #GENERATION POUR BETA SUIVANT UNE LOI UNIFORME
        DESSERT, CHOIX, VALEURS, PROBA, INFLUENCE = ESSAI(B_SIMULATION, A_SIMULATION, 0, modele = MODELE)

    #CONSERVATION DES DONNEES
        BANQUE = {'BETA': [], 'ALPHA': [], 'LL': [], 'LL_ALPHA': []}

    #RECHERCHE BETA
        for TEST in RECHERCHE:
            BETA, LL = LIKELIHOOD(DESSERT,CHOIX, VALEURS, PROBA, BETA = TEST, INFLUENCE = INFLUENCE, modele = 'linear', option = 'beta' )
            BANQUE['BETA'].append(BETA)
            BANQUE['LL'].append(LL)
    
        LL_max = max(BANQUE['LL'])
        maximum = BANQUE['BETA'][BANQUE['LL'].index(LL_max)]

        for TEST in RECHERCHE:
            ALPHA1, LL_ALPHA = LIKELIHOOD(DESSERT,CHOIX, VALEURS, PROBA, BETA = maximum, ALPHA = TEST, INFLUENCE = INFLUENCE, modele = 'linear',option = 'alpha' )
            BANQUE['ALPHA'].append(ALPHA1)
            BANQUE['LL_ALPHA'].append(LL_ALPHA)
            
        return BANQUE, B_SIMULATION, A_SIMULATION



    elif MODELE == 'triggering':

      #GENERATION POUR BETA SUIVANT UNE LOI UNIFORME
        DESSERT, CHOIX, VALEURS, PROBA, INFLUENCE = ESSAI(B_SIMULATION, 0 , D_SIMULATION, modele = MODELE)

      #CONSERVATION DES DONNEES
        BANQUE = {'BETA': [], 'D': [], 'LL': [], 'LL_D': []}

      #RECHERCHE BETA
        for TEST in RECHERCHE:
            BETA, LL = LIKELIHOOD(DESSERT,CHOIX, VALEURS, PROBA, BETA = TEST, INFLUENCE = INFLUENCE, modele = MODELE, option = 'beta' )
            BANQUE['BETA'].append(BETA)
            BANQUE['LL'].append(LL)
      
        LL_max = max(BANQUE['LL'])
        maximum = BANQUE['BETA'][BANQUE['LL'].index(LL_max)]

        for TEST in RECHERCHE_D:
            TRIGGER, LL_D = LIKELIHOOD(DESSERT,CHOIX, VALEURS, PROBA, BETA = maximum, D = TEST, INFLUENCE = INFLUENCE, modele = MODELE,option = MODELE )
            BANQUE['D'].append(TRIGGER)
            BANQUE['LL_D'].append(LL_D)
            
        return BANQUE, B_SIMULATION, D_SIMULATION
```

```python
T = 1000
BETA = []
ALPHA = []
D = []

for i in range(T):
    BETA.append(np.random.uniform(low=0, high=5, size=1))
    ALPHA.append(np.random.uniform(low=0, high=5, size=1))
    D.append(np.random.uniform(low=0, high=20, size=1))
```

```python
def confusion_no(B_SIMULATION = 0, A_SIMULATION = 0, D_SIMULATION = 0, MODELE = 'no'):
#PARAMETRE ENTREE : PARAMETRES DE LA SIMULATION
#PARAMETRE ENTREE : MODELE -> CHECKING MODELE

#ATTENTION CONFUSION_NO SIMULE UNE PERSONNE SANS INFLUENCE ET ON CHECK LE MODELE SELON LE PARAMETRE MODELE ('no','linear','triggering')
    
    
    if MODELE == 'no':
    #GENERATION POUR BETA SUIVANT UNE LOI UNIFORME
        DESSERT, CHOIX, VALEURS, PROBA, INFLUENCE = ESSAI(B_SIMULATION, 0, 0, modele = MODELE)

    #CONSERVATION DES DONNEES
        BANQUE = {'BETA': [], 'LL': []}
    #RECHERCHE
        for TEST in RECHERCHE:
            BETA, LL = LIKELIHOOD(DESSERT,CHOIX, VALEURS, PROBA, TEST )
            BANQUE['BETA'].append(BETA)
            BANQUE['LL'].append(LL)

        return BANQUE, B_SIMULATION

    elif MODELE == 'linear':

    #GENERATION POUR BETA SUIVANT UNE LOI UNIFORME
        DESSERT, CHOIX, VALEURS, PROBA, INFLUENCE = ESSAI(B_SIMULATION, 0, 0, modele = MODELE)

    #CONSERVATION DES DONNEES
        BANQUE = {'BETA': [], 'ALPHA': [], 'LL': [], 'LL_ALPHA': []}

    #RECHERCHE BETA
        for TEST in RECHERCHE:
            BETA, LL = LIKELIHOOD(DESSERT,CHOIX, VALEURS, PROBA, BETA = TEST, INFLUENCE = INFLUENCE, modele = 'linear', option = 'beta' )
            BANQUE['BETA'].append(BETA)
            BANQUE['LL'].append(LL)
    
        LL_max = max(BANQUE['LL'])
        maximum = BANQUE['BETA'][BANQUE['LL'].index(LL_max)]

        for TEST in RECHERCHE:
            ALPHA1, LL_ALPHA = LIKELIHOOD(DESSERT,CHOIX, VALEURS, PROBA, BETA = maximum, ALPHA = TEST, INFLUENCE = INFLUENCE, modele = 'linear',option = 'alpha' )
            BANQUE['ALPHA'].append(ALPHA1)
            BANQUE['LL_ALPHA'].append(LL_ALPHA)
        return BANQUE, B_SIMULATION, A_SIMULATION



    elif MODELE == 'triggering':

      #GENERATION POUR BETA SUIVANT UNE LOI UNIFORME
        DESSERT, CHOIX, VALEURS, PROBA, INFLUENCE = ESSAI(B_SIMULATION, 0, 0, modele = MODELE)

      #CONSERVATION DES DONNEES
        BANQUE = {'BETA': [], 'D': [], 'LL': [], 'LL_D': []}

      #RECHERCHE BETA
        for TEST in RECHERCHE:
            BETA, LL = LIKELIHOOD(DESSERT,CHOIX, VALEURS, PROBA, BETA = TEST, INFLUENCE = INFLUENCE, modele = MODELE, option = 'beta' )
            BANQUE['BETA'].append(BETA)
            BANQUE['LL'].append(LL)
      
        LL_max = max(BANQUE['LL'])
        maximum = BANQUE['BETA'][BANQUE['LL'].index(LL_max)]

        for TEST in RECHERCHE_D:
            TRIGGER, LL_D = LIKELIHOOD(DESSERT,CHOIX, VALEURS, PROBA, BETA = maximum, D = TEST, INFLUENCE = INFLUENCE, modele = MODELE,option = MODELE )
            BANQUE['D'].append(TRIGGER)
            BANQUE['LL_D'].append(LL_D)
        return BANQUE, B_SIMULATION, D_SIMULATION
```

```python
def confusion_linear(B_SIMULATION = 0, A_SIMULATION = 0, D_SIMULATION = 0, MODELE = 'linear'):

    if MODELE == 'no':
    #GENERATION POUR BETA SUIVANT UNE LOI UNIFORME
        DESSERT, CHOIX, VALEURS, PROBA, INFLUENCE = ESSAI(B_SIMULATION, A_SIMULATION, 0, modele = MODELE)

    #CONSERVATION DES DONNEES
        BANQUE = {'BETA': [], 'LL': []}
    #RECHERCHE
        for TEST in RECHERCHE:
            BETA, LL = LIKELIHOOD(DESSERT,CHOIX, VALEURS, PROBA, TEST )
            BANQUE['BETA'].append(BETA)
            BANQUE['LL'].append(LL)

        return BANQUE, B_SIMULATION

    elif MODELE == 'linear':

    #GENERATION POUR BETA SUIVANT UNE LOI UNIFORME
        DESSERT, CHOIX, VALEURS, PROBA, INFLUENCE = ESSAI(B_SIMULATION, A_SIMULATION, 0, modele = MODELE)

    #CONSERVATION DES DONNEES
        BANQUE = {'BETA': [], 'ALPHA': [], 'LL': [], 'LL_ALPHA': []}

    #RECHERCHE BETA
        for TEST in RECHERCHE:
            BETA, LL = LIKELIHOOD(DESSERT,CHOIX, VALEURS, PROBA, BETA = TEST, INFLUENCE = INFLUENCE, modele = 'linear', option = 'beta' )
            BANQUE['BETA'].append(BETA)
            BANQUE['LL'].append(LL)
    
        LL_max = max(BANQUE['LL'])
        maximum = BANQUE['BETA'][BANQUE['LL'].index(LL_max)]

        for TEST in RECHERCHE:
            ALPHA1, LL_ALPHA = LIKELIHOOD(DESSERT,CHOIX, VALEURS, PROBA, BETA = maximum, ALPHA = TEST, INFLUENCE = INFLUENCE, modele = 'linear',option = 'alpha' )
            BANQUE['ALPHA'].append(ALPHA1)
            BANQUE['LL_ALPHA'].append(LL_ALPHA)
        return BANQUE, B_SIMULATION, A_SIMULATION



    elif MODELE == 'triggering':

      #GENERATION POUR BETA SUIVANT UNE LOI UNIFORME
        DESSERT, CHOIX, VALEURS, PROBA, INFLUENCE = ESSAI(B_SIMULATION, A_SIMULATION, 0, modele = MODELE)

      #CONSERVATION DES DONNEES
        BANQUE = {'BETA': [], 'D': [], 'LL': [], 'LL_D': []}

      #RECHERCHE BETA
        for TEST in RECHERCHE:
            BETA, LL = LIKELIHOOD(DESSERT,CHOIX, VALEURS, PROBA, BETA = TEST, INFLUENCE = INFLUENCE, modele = MODELE, option = 'beta' )
            BANQUE['BETA'].append(BETA)
            BANQUE['LL'].append(LL)
      
            LL_max = max(BANQUE['LL'])
            maximum = BANQUE['BETA'][BANQUE['LL'].index(LL_max)]

        for TEST in RECHERCHE:
            TRIGGER, LL_D = LIKELIHOOD(DESSERT,CHOIX, VALEURS, PROBA, BETA = maximum, D = TEST, INFLUENCE = INFLUENCE, modele = MODELE,option = MODELE )
            BANQUE['D'].append(TRIGGER)
            BANQUE['LL_D'].append(LL_D)
        return BANQUE, B_SIMULATION, D_SIMULATION
```

```python
def confusion_triggering(B_SIMULATION = 0, A_SIMULATION = 0, D_SIMULATION = 0, MODELE = 'triggering'):

    if MODELE == 'no':
    #GENERATION POUR BETA SUIVANT UNE LOI UNIFORME
        DESSERT, CHOIX, VALEURS, PROBA, INFLUENCE = ESSAI(B_SIMULATION, 0 , D_SIMULATION, modele = MODELE)
    #CONSERVATION DES DONNEES
        BANQUE = {'BETA': [], 'LL': []}
    #RECHERCHE
        for TEST in RECHERCHE:
            BETA, LL = LIKELIHOOD(DESSERT,CHOIX, VALEURS, PROBA, TEST )
            BANQUE['BETA'].append(BETA)
            BANQUE['LL'].append(LL)

        return BANQUE, B_SIMULATION

    elif MODELE == 'linear':

    #GENERATION POUR BETA SUIVANT UNE LOI UNIFORME
        DESSERT, CHOIX, VALEURS, PROBA, INFLUENCE = ESSAI(B_SIMULATION, 0 , D_SIMULATION, modele = MODELE)

    #CONSERVATION DES DONNEES
        BANQUE = {'BETA': [], 'ALPHA': [], 'LL': [], 'LL_ALPHA': []}

    #RECHERCHE BETA
        for TEST in RECHERCHE:
            BETA, LL = LIKELIHOOD(DESSERT,CHOIX, VALEURS, PROBA, BETA = TEST, INFLUENCE = INFLUENCE, modele = 'linear', option = 'beta' )
            BANQUE['BETA'].append(BETA)
            BANQUE['LL'].append(LL)
    
        LL_max = max(BANQUE['LL'])
        maximum = BANQUE['BETA'][BANQUE['LL'].index(LL_max)]

        for TEST in RECHERCHE:
            ALPHA1, LL_ALPHA = LIKELIHOOD(DESSERT,CHOIX, VALEURS, PROBA, BETA = maximum, ALPHA = TEST, INFLUENCE = INFLUENCE, modele = 'linear',option = 'alpha' )
            BANQUE['ALPHA'].append(ALPHA1)
            BANQUE['LL_ALPHA'].append(LL_ALPHA)
        return BANQUE, B_SIMULATION, A_SIMULATION



    elif MODELE == 'triggering':

      #GENERATION POUR BETA SUIVANT UNE LOI UNIFORME
        DESSERT, CHOIX, VALEURS, PROBA, INFLUENCE = ESSAI(B_SIMULATION, 0 , D_SIMULATION, modele = MODELE)

      #CONSERVATION DES DONNEES
        BANQUE = {'BETA': [], 'D': [], 'LL': [], 'LL_D': []}

      #RECHERCHE BETA
        for TEST in RECHERCHE:
            BETA, LL = LIKELIHOOD(DESSERT,CHOIX, VALEURS, PROBA, BETA = TEST, INFLUENCE = INFLUENCE, modele = MODELE, option = 'beta' )
            BANQUE['BETA'].append(BETA)
            BANQUE['LL'].append(LL)
      
        LL_max = max(BANQUE['LL'])
        maximum = BANQUE['BETA'][BANQUE['LL'].index(LL_max)]

        for TEST in RECHERCHE:
            TRIGGER, LL_D = LIKELIHOOD(DESSERT,CHOIX, VALEURS, PROBA, BETA = maximum, D = TEST, INFLUENCE = INFLUENCE, modele = MODELE,option = MODELE )
            BANQUE['D'].append(TRIGGER)
            BANQUE['LL_D'].append(LL_D)
        return BANQUE, B_SIMULATION, D_SIMULATION
```

```python
MATRICE = {'index' : ['no','linear','triggering'], 'no' : [], 'linear' : [], 'triggering' : []}

round = -1
for name in MATRICE['index']:
    round = round + 1
    print('Itération N°',round)
    if name == 'no':
        for i in range(T):
            print('Sous itération N°',i)
            BANQUE, B_SIMULATION1 = confusion_no(B_SIMULATION = BETA[i][0], MODELE = 'no')
            MATRICE['no'].append(max(BANQUE['LL']))
            BANQUE, B_SIMULATION2, A_SIMULATION2 = confusion_no(B_SIMULATION = BETA[i][0], MODELE = 'linear')
            MATRICE['linear'].append(max(BANQUE['LL_ALPHA']))
            BANQUE, B_SIMULATION3, D_SIMULATION3 = confusion_no(B_SIMULATION = BETA[i][0], MODELE = 'triggering')
            MATRICE['triggering'].append(max(BANQUE['LL_D']))

    elif name == 'linear':
        for i in range(T):
            print('Sous itération N°',i)
            BANQUE, B_SIMULATION1 = confusion_linear(B_SIMULATION = BETA[i][0], A_SIMULATION = ALPHA[i][0], MODELE = 'no')
            MATRICE['no'].append(max(BANQUE['LL']))
            BANQUE, B_SIMULATION2, A_SIMULATION2 = confusion_linear(B_SIMULATION = BETA[i][0], A_SIMULATION = ALPHA[i][0], MODELE = 'linear')
            MATRICE['linear'].append(max(BANQUE['LL_ALPHA']))
            BANQUE, B_SIMULATION3, D_SIMULATION3 = confusion_linear(B_SIMULATION = BETA[i][0], A_SIMULATION = ALPHA[i][0], MODELE = 'triggering')
            MATRICE['triggering'].append(max(BANQUE['LL_D']))

    elif name == 'triggering':
        for i in range(T):
            print('Sous itération N°',i)
            BANQUE, B_SIMULATION1 = confusion_triggering(B_SIMULATION = BETA[i][0], D_SIMULATION = D[i][0], MODELE = 'no')
            MATRICE['no'].append(max(BANQUE['LL']))
            BANQUE, B_SIMULATION2, A_SIMULATION2 = confusion_triggering(B_SIMULATION = BETA[i][0], D_SIMULATION = D[i][0], MODELE = 'linear')
            MATRICE['linear'].append(max(BANQUE['LL_ALPHA']))
            BANQUE, B_SIMULATION3, D_SIMULATION3 = confusion_triggering(B_SIMULATION = BETA[i][0], D_SIMULATION = D[i][0], MODELE = 'triggering')
            MATRICE['triggering'].append(max(BANQUE['LL_D']))
```
